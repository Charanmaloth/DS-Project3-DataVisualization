```python
# Reshape the data from wide to long format
df_population_long = cp.melt(id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
                                        var_name="Year", value_name="Population")
df_fertility_long = fr.melt(id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
                                      var_name="Year", value_name="Fertility Rate")
df_life_expectancy_long = le.melt(id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
                                                  var_name="Year", value_name="Life Expectancy")
```

```python
# Convert 'Year' column to integers
df_population_long['Year'] = df_population_long['Year'].astype(int)
df_fertility_long['Year'] = df_fertility_long['Year'].astype(int)
df_life_expectancy_long['Year'] = df_life_expectancy_long['Year'].astype(int)
```

```python
# Drop unnecessary columns
df_population_long = df_population_long.drop(columns=["Indicator Name", "Indicator Code"])
df_fertility_long = df_fertility_long.drop(columns=["Indicator Name", "Indicator Code"])
df_life_expectancy_long = df_life_expectancy_long.drop(columns=["Indicator Name", "Indicator Code"])
```

```python
# Display the reshaped data
print(df_population_long.head())
print(df_fertility_long.head())
print(df_life_expectancy_long.head())
```

```python
# Merge the dataframes on 'Country Name' and 'Year'
merged_df = pd.merge(df_population_long, df_fertility_long, on=["Country Name", "Country Code", "Year"])
merged_df = pd.merge(merged_df, df_life_expectancy_long, on=["Country Name", "Country Code", "Year"])

# Display the merged DataFrame
print(merged_df.head())
```

```python
# Check for missing values
print(merged_df.isnull().sum())

# Fill missing values or handle them as appropriate
merged_df.fillna(method='ffill', inplace=True)

# Verify no missing values
print(merged_df.isnull().sum())
```

```python
#Check existence of metadata
if 'Metadata_Country.csv' in locals():
    merged_df = pd.merge(merged_df, meta, on=["Country Name", "Country Code"])
```

```python
print(merged_df.head())
print(merged_df.info())
```

```python
#Population Trends
population_trend = merged_df.groupby('Year')['Population'].sum().reset_index()
plt.figure(figsize=(12, 8))
plt.plot(population_trend['Year'], population_trend['Population'], marker='o', linestyle='-')
plt.title('Population Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot the histogram(Fertility rate distribution)
import seaborn as sns
plt.figure(figsize=(12, 8))
df_1960 = df_fertility_long[df_fertility_long['Year'] == 1960]
sns.histplot(df_1960['Fertility Rate'], bins=10, edgecolor='black')
plt.xlabel('Fertility Rate')
plt.ylabel('Countries')
plt.title("Fertility Rate Distribution")
plt.show()
```

```python
#Life expectancy variation
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.boxplot(data=merged_df, x='Year', y='Life Expectancy')
plt.title('Life Expectancy Variation Over Years')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Correlation Analysis (scatter plot)
plt.figure(figsize=(12, 8))
sns.scatterplot(data=merged_df, x='Fertility Rate', y='Life Expectancy')
plt.title('Correlation Analysis')
plt.xlabel('Fertility Rate')
plt.ylabel('Life Expectancy')
plt.tight_layout()
plt.show()
```
